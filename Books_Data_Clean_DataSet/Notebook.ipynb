```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
df=pd.read_csv("Books_Data_Clean.csv")
```

```python
df.head()
```

```python
df.describe()
```

describe() metodu, bir DataFrame içindeki sayısal (veya belirli parametreler verilirse kategorik) sütunlar için aşağıdaki betimleyici istatistikleri sağlar:

count: Her sütundaki geçerli (NaN olmayan) değerlerin sayısı.
mean: Her sütundaki değerlerin ortalaması.
std: Her sütundaki değerlerin standart sapması.
min: Her sütundaki minimum değer.
25%: Her sütundaki değerlerin yüzde 25'lik dilimi (1. çeyrek).
50% (medyan): Her sütundaki değerlerin yüzde 50'lik dilimi (2. çeyrek).
75%: Her sütundaki değerlerin yüzde 75'lik dilimi (3. çeyrek).
max: Her sütundaki maksimum değer.

```python
df=df[df["Publishing Year"] > 1900]
```

```python
df.isna().sum()
```

```python
df.dropna(subset="Book Name",inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

```python
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distrubition of Publishing Year")
plt.show()
```

```python
df["genre"].value_counts().plot(kind="bar")
plt.xlabel("Genre")
plt.ylabel("Number of Book")
plt.title("Number of books in Each Genre")
plt.show()
```

```python
df.groupby("Author")["Book_average_rating"].mean()
```

```python
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending=False)
```

```python
sbn.boxplot(x="genre", y="Book_ratings_count",data=df)
plt.xlabel("Genre")
plt.ylabel("Book Ratings Count")
plt.title("Box plot of Books Ratings Count for Each Genre")
plt.show()
```

```python
plt.scatter(df["sale price"],df["units sold"])
plt.xlabel("Sale price")
plt.ylabel("Units sold")
plt.title("Scatter plot of Sale price vs Units sold")
plt.show()
```

```python
language_counts=df["language_code"].value_counts()
```

```python
plt.pie(language_counts,labels=language_counts.index,startangle=90,autopct="%1.1f%%")
plt.title("Language Distrubiton of Books")
plt.show()
```

```python
df.groupby("Publisher ")["publisher revenue"].sum().sort_values(ascending=False)
```

```python
df.groupby("Author_Rating")["Book_ratings_count"].mean().sort_values(ascending=False)
```

```python
df.groupby("language_code").size()
```

```python
plt.scatter(df["Book_average_rating"], df["Book_ratings_count"])
plt.xlabel("Book avarage rating")
plt.ylabel("Book rating count")
plt.title("Scatter plot of Book avarage rating vs Book ratings count")
-plt.show()
```

```python
total_gross_sales_by_author=df.groupby("Author")["gross sales"].sum()
```

```python
total_gross_sales_by_author.sort_values(ascending=False).head(10).plot(kind="bar")
plt.xlabel("Author")
plt.ylabel("Total gross sales")
plt.title("Total gross sales for each Author")
plt.show()
```

```python
df.groupby("Publishing Year")["units sold"].sum().plot(kind="line", marker="o")
plt.xlabel("Publishing Year")
plt.ylabel("Total units Sold")
plt.title("Total units sold over the year")
plt.show()
```
